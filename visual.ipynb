```python
car_sales["Price"] = car_sales["Price"].str.replace('[\$\,\.]', '', regex=True)
car_sales["Price"] = car_sales["Price"].str[:-2]
car_sales
```

```python
car_sales["Date_range"] = pd.date_range("1/1/2020", periods=len(car_sales))
car_sales
```

```python
car_sales["Total_sales"] = car_sales["Price"].astype(int).cumsum()
car_sales
```

```python

car_sales.plot(x="Date_range", y="Total_sales")
```

<h1 style="color:80ff0ff0080">why one should i use pyplot method or object oriented method</h1>


```python
over_50 = h_d[h_d["age"] > 50]
over_50
```

```python
over_50.plot(kind="scatter", x="age", y="chol", c="target")
```

```python
fig, axs = plt.subplots(figsize=(10, 6))
over_50.plot(kind="scatter", x="age", y="chol", c="target", ax=axs)
axs.set_xlim([45, 100])
```

```python
fig, axs = plt.subplots()
scatter = axs.scatter(x=over_50["age"], y=over_50["chol"], c=over_50["target"])
axs.set(title="heart disease of patients", xlabel="age", ylabel="cholestrol")
axs.legend(*scatter.legend_elements(), title="Target")
axs.axhline(over_50["chol"].mean(), linestyle="--")
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
ax.plot([0, 10], [0, 10])

# Shade between y=2 and y=5 across the full width
ax.axhspan(2, 5, color='red', alpha=1)

plt.show()
```

```python
fig, (ax0, ax1) = plt.subplots(nrows=2, ncols=1, figsize=(10, 10), sharex=True)
scatter = ax0.scatter(x=over_50["age"], y=over_50["chol"], c=over_50["target"])
ax0.set(title="heart disease and cholestrol levels",
        xlabel="age", ylabel="cholestrol")
ax0.axhline(over_50["chol"].mean(), linestyle="--")
ax0.legend(*scatter.legend_elements(), title="Target")

scatter = ax1.scatter(
    x=over_50["age"], y=over_50["thalach"], c=over_50["target"])
ax1.set(title="heart disease and thalach level",
        xlabel="age", ylabel="thalach")
ax1.legend(*scatter.legend_elements(), title="Target")
ax1.axhline(over_50["thalach"].mean(), linestyle='--')

fig.suptitle("heart disease analysis", fontsize=16,
             fontstyle="italic", fontweight="bold")
```

```python
import matplotlib.patches as patches
fig, (ax0, ax1) = plt.subplots(nrows=2, ncols=1, figsize=(10, 10), sharex=True)

scatter = ax0.scatter(x=over_50["age"], y=over_50["chol"], c=over_50["target"])
ax0.set(title="Heart disease and cholesterol levels",
        xlabel="Age", ylabel="Cholesterol")
ax0.axhline(over_50["chol"].mean(), linestyle="--")
ax0.legend(*scatter.legend_elements(), title="Target")

scatter = ax1.scatter(
    x=over_50["age"], y=over_50["thalach"], c=over_50["target"])
ax1.set(title="Heart disease and thalach level",
        xlabel="Age", ylabel="Thalach")
ax1.legend(*scatter.legend_elements(), title="Target")
ax1.axhline(over_50["thalach"].mean(), linestyle="--")

fig.suptitle("Heart disease analysis", fontsize=16,
             fontstyle="italic", fontweight="bold")

# ✅ Correct import

# ✅ Draw rectangle in figure coordinates
rect = patches.Rectangle(
    (0.05, 0.05),   # bottom-left corner (x, y) in figure coords
    0.9, 0.9,       # width, height in figure coords
    linewidth=2,
    edgecolor='red',
    facecolor='none'
)

# ✅ Correct way to add to figure
fig.add_artist(rect)

plt.show()
```
